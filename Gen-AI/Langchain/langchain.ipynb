Basic usage of llm to generate answers from a question with OpenAI GPT model

```python
from langchain.llms import OpenAI

llm = OpenAI(temperature=0.6)

text = "What are 5 vacation destinations for someone who likes to eat pasta?"
print(llm(text))
```

Usage of Promt Template: No need to give entire sentence as an input

```python
from langchain.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["food"],
    template="What are 5 vacation destinations for someone who likes to eat {food}?",
)

prompt.format(food="dessert")
```

Usage of LLM Chain is two combine the PromtTemplate with LLM

```python
from langchain.chains import LLMChain
chain = LLMChain(llm=llm, prompt=prompt)

print(chain.run("fruit"))
```

Sequential Chain: Used to run two promt template.
Input -> First LLM Chain -> Output as Input -> Second LLM Chain -> Output

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="I want to open a restaurant for {cuisine} food. Suggest a fency name for this."
)

name_chain = LLMChain(llm=llm, prompt=prompt_template_name)

prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}."
)

food_items_chain = LLMChain(llm=llm, prompt=prompt_template_items)

from langchain.chains import SimpleSequentialChain
chain = SimpleSequentialChain(chains=[name_chain, food_items_chain])

print(chain.run("Indian"))
```

The Problem with Simple Sequential Chain is we are only getting one/end output but if we want to get each chains output then we've to use Sequential Chain instead.

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="I want to open a restaurant for {cuisine} food. Suggest a fency name for this.",
    output_key="restaurant_name"
)

name_chain = LLMChain(llm=llm, prompt=prompt_template_name)

prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}.",
    output_key="menu_items"
)

food_items_chain = LLMChain(llm=llm, prompt=prompt_template_items)

from langchain.chains import SequentialChain

chain = SequentialChain(
    chains=[name_chain, food_items_chain],
    input_variables=['cuisine'],
    output_variables=['restaurant_name', 'menu_items']
)

print(chain({'cuisine': 'Indian'}))
```

Agents are a coworker works between LLM and Request like if any data isnt stored in LLM then it simply request it through internet like Wikipedia or Google and get the information....

```python
#import agents

from langchain.agents import load_tools, AgentType, initialize_agent

tools = load_tools('wikipedia', 'llm-math', llm=llm)

agent = initialize_agent(AgentType.ZERO_SHOT_REACT_DESCRIPTION, tools=tools)

agent.run("What is the current GDP of India?")
```

Memory are used to save earliar responses into buffers of langchain and can be stored further into Databases just like OpenAI stored chats...

```python
from langchain.memory import ConversationalBufferMemory

memory = ConversationalBufferMemory()

chain = LLMChain(llm=llm, prompt=prompt, memory=memory)

name = chain.run("Italian")

print(name)

print(chain.memory)
```
